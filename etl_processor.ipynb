```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
# upload  anime csv file
csv_file = "./Resources/Anime.csv"
anime_df = pd.read_csv(csv_file)
anime_df.head()
```

```python
# upload  netflix csv file
csv_file = "./Resources/NetflixOriginals.csv"
netflixog_df = pd.read_csv(csv_file)
netflixog_df.head()
```

```python
# upload US/Youtube csv file
csv_file = "./Resources/USvideos.csv"
usvideos_df = pd.read_csv(csv_file)
usvideos_df.head()
```

```python
# connect to local database 
rds_connection_string = "postgres:postgres@localhost:5432/media"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
# use pandas to load csv converted DF into DB
anime_df.to_sql(name='Anime', con=engine, if_exists='append', index=False)
netflixog_df.to_sql(name='Netflix', con=engine, if_exists='append', index=False)
usvideos_df.to_sql(name='Youtube/US', con=engine, if_exists='append', index=False)
```
